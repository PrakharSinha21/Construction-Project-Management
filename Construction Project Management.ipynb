```python
df_task['Priority'].value_counts()
```

```python
#Behavioural Failure is the most applied in the Priority column. 
#The tasks associated with this category relate to Safety.
```

```python
df_task.loc[df_task['Task Group']=='Safety','Priority']=df_task['Priority'].fillna('Behavioural Failure')
df_task.loc[df_task['Task Group']=='Safety']
```

```python
#the remaining rows can be filled with 'Other'.

df_task['Priority']=df_task['Priority'].fillna('others')
df_task['Priority']
```

```python
df_task['Target'].value_counts()
```

```python
df_task['Target'].mean()
```

```python
#we will use 'mean' value to fill in the missing data.

df_task['Target']=df_task['Target'].fillna(df_task['Target'].mean())
df_task['Target']
```

```python
df_task['Association'].value_counts()
```

```python
#Now, let's take a look at the column named, Association.
#It is unclear what this column is for, however the values in this column include 'FormAnswer'.
#For this reason, it would seem that the data is missing because it hasn't been recorded.
#We will fill the missing values with 'FormAnswer'

df_task['Association']=df_task['Association'].fillna('FormAnswer')
df_task['Association']
```

```python
df_task['Cause'].value_counts()
```

```python
#The column named, Cause also contains some missing values. They seem to have not been recorded.
#Existing values in this column include 'Safety documentation', ' House keeping', or 'plant and equipment'.
#It's fair to say that the missing values have simply not been recorded and similarly,
#we could fill them with 'NoRecord'.

df_task['Cause']=df_task['Cause'].fillna('NoRecord')
df_task['Cause']

```

```python
df_task['To Package'].value_counts()
```

```python
#The column named, To Package indicates which trade contractor or department the task is associated with.
#By looking at the full dataset, it is clear that the missing values are for tasks associated with design.
#Therefore, it would be fair to say the data has not been recorded rather than it doesn't exist.
#We can fill the missing values with 'Design'.

df_task['To Package']=df_task['To Package'].fillna('Design')
df_task['To Package']
```

```python
#The columns headed 'Images', 'Comments' and 'Documents' contain Boolean data types.
#There's approx. 5% or less of missing data in these columns.
#Rather than assume what the value should be, we will fill the missing values with 'N/A'

df_task[['Images','Comments','Documents']]=df_task[['Images','Comments','Documents']].fillna('N/A')
df_task[['Images','Comments','Documents']]
```

```python
df_task['Task Group'].value_counts()
```

```python
#Finally, the column named, Task Group has 50 missing values.
#This column details which sub-group the task is associated with.
#Given that Safety has been assigned the most in this column,
#we will fill any missing values with the sub-category, "Safety".

df_task['Task Group']=df_task['Task Group'].fillna('Safety')
df_task['Task Group']
```

```python
subset_df_tasks = df_task[['Ref', 'Status', 'Location', 'Description', 'Created', 'Type', 'To Package', 'Status Changed', 'Priority', 'Cause', 'project', 'Report Status', 'Task Group']]
subset_df_tasks.dtypes
```

```python
subset_df_tasks.shape
```

```python
#filter tasks by project and then analyse

data_by_project=subset_df_tasks.drop(columns=['Status', 'Location', 'Description', 'Created', 'Type',
                                              'To Package', 'Status Changed', 'Priority', 'Cause',
                                              'Report Status', 'Task Group'])
```

```python
data_by_project=data_by_project.groupby('project').count().rename(columns={'Ref':'Quantity'}).reset_index()
data_by_project
```

```python
plt.figure(figsize=(9,8))

plt.pie(data_by_project ['Quantity'], labels=data_by_project['project'], autopct='%1.1f%%')

plt.legend(title='Project Ref', loc='upper right', bbox_to_anchor=(1.3, 1))
plt.title('Tasks by Project',fontweight='bold')

plt.show()
```

```python
df_package=subset_df_tasks['To Package'].value_counts().reset_index()
df_package.columns=['Trade Package','Quantity']
df_package
```
